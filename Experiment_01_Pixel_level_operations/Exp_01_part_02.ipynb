# Algorithm: Piecewise Linear Transformation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = np.arange(256)
s = np.zeros(256)
```

```python
r1 , r2 = 70, 140
s1 , s2 = 30 , 200 
```

```python
for i in range(256):
    if r[i] < r1:
        s[i] = (s1/r1) * r[i]
    elif r[i] <= r2:
        s[i] = ((s2 - s1)/(r2 - r1)) * (r[i] - r1) + s1
    else:
        s[i] = ((255 - s2)/(255 - r2)) * (r[i] - r2) + s2
```

```python
plt.plot(r, s)
plt.xlabel("Input gray level (r)")
plt.ylabel("Output gray level (s)")
plt.title("Piecewise Linear Transformation (Contarst streaching)")
plt.grid()
plt.show()
```
