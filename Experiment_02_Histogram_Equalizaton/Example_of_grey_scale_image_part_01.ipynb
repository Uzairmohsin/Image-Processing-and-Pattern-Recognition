# PART-2: Histogram Equalization on RGB Components

```python
plt.figure(figsize=(5,5))
plt.imshow(img_rgb)
plt.title("Original RGB Image")
plt.axis("off")
plt.show()
```

```python
R, G, B = cv2.split(img_rgb)
```

```python
R_eq = cv2.equalizeHist(R)
G_eq = cv2.equalizeHist(G)
B_eq = cv2.equalizeHist(B)
```

```python
color_eq = cv2.merge((R_eq, G_eq, B_eq))
```

```python
plt.figure(figsize=(12, 8))

# ---------- Original RGB Image ----------
plt.subplot(2, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Color Image")
plt.axis("off")

# ---------- RGB Equalized Image ----------
plt.subplot(2, 2, 2)
plt.imshow(color_eq)
plt.title("RGB Histogram Equalized Image")
plt.axis("off")

# ---------- Histogram of Original RGB ----------
plt.subplot(2, 2, 3)
colors = ('r', 'g', 'b')
labels = ('Red', 'Green', 'Blue')

for i, col in enumerate(colors):
    hist = cv2.calcHist([img_rgb], [i], None, [256], [0,256])
    plt.plot(hist, color=col, label=labels[i])

plt.title("Histogram of Original RGB Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.legend()

# ---------- Histogram of RGB Equalized ----------
plt.subplot(2, 2, 4)

for i, col in enumerate(colors):
    hist = cv2.calcHist([color_eq], [i], None, [256], [0,256])
    plt.plot(hist, color=col, label=labels[i])

plt.title("Histogram of RGB Equalized Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 12))
plot_index = 1

# ---- RED CHANNEL ----
plt.subplot(3,4,plot_index)
plt.imshow(R, cmap='gray')
plt.title("Red Channel - Original")
plt.axis("off")
plot_index += 1

plt.subplot(3,4,plot_index)
plt.imshow(R_eq, cmap='gray')
plt.title("Red Channel - Equalized")
plt.axis("off")
plot_index += 1

plt.subplot(3,4,plot_index)
plt.hist(R.ravel(), bins=256, color='red')
plt.title("Red Histogram (Original)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plot_index += 1

plt.subplot(3,4,plot_index)
plt.hist(R_eq.ravel(), bins=256, color='red')
plt.title("Red Histogram (Equalized)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plot_index += 1

# ---- GREEN CHANNEL ----
plt.subplot(3,4,plot_index)
plt.imshow(G, cmap='gray')
plt.title("Green Channel - Original")
plt.axis("off")
plot_index += 1

plt.subplot(3,4,plot_index)
plt.imshow(G_eq, cmap='gray')
plt.title("Green Channel - Equalized")
plt.axis("off")
plot_index += 1

plt.subplot(3,4,plot_index)
plt.hist(G.ravel(), bins=256, color='green')
plt.title("Green Histogram (Original)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plot_index += 1

plt.subplot(3,4,plot_index)
plt.hist(G_eq.ravel(), bins=256, color='green')
plt.title("Green Histogram (Equalized)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plot_index += 1

# ---- BLUE CHANNEL ----
plt.subplot(3,4,plot_index)
plt.imshow(B, cmap='gray')
plt.title("Blue Channel - Original")
plt.axis("off")
plot_index += 1

plt.subplot(3,4,plot_index)
plt.imshow(B_eq, cmap='gray')
plt.title("Blue Channel - Equalized")
plt.axis("off")
plot_index += 1

plt.subplot(3,4,plot_index)
plt.hist(B.ravel(), bins=256, color='blue')
plt.title("Blue Histogram (Original)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plot_index += 1

plt.subplot(3,4,plot_index)
plt.hist(B_eq.ravel(), bins=256, color='blue')
plt.title("Blue Histogram (Equalized)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plot_index += 1

plt.tight_layout()
plt.show()
```

```python
hsv_img = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
H, S, V = cv2.split(hsv_img)
```

```python
V_eq = cv2.equalizeHist(V)
```

```python
hsv_eq = cv2.merge((H, S, V_eq))
```

```python
rgb_hsi_eq = cv2.cvtColor(hsv_eq, cv2.COLOR_HSV2RGB)
```

```python
plt.figure(figsize=(15, 12))

# ---------- IMAGES ----------
plt.subplot(4,2,1)
plt.imshow(img_rgb)
plt.title("Original RGB Image")
plt.axis("off")

plt.subplot(4,2,2)
plt.imshow(rgb_hsi_eq)
plt.title("HSI Intensity Equalized Image")
plt.axis("off")

# ---------- ORIGINAL IMAGE HISTOGRAMS ----------
plt.subplot(4,2,3)
plt.hist(img_rgb[:,:,0].ravel(), bins=256, color='red')
plt.title("Original - Red Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.subplot(4,2,5)
plt.hist(img_rgb[:,:,1].ravel(), bins=256, color='green')
plt.title("Original - Green Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.subplot(4,2,7)
plt.hist(img_rgb[:,:,2].ravel(), bins=256, color='blue')
plt.title("Original - Blue Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

# ---------- HSI IMAGE HISTOGRAMS ----------
plt.subplot(4,2,4)
plt.hist(rgb_hsi_eq[:,:,0].ravel(), bins=256, color='red')
plt.title("HSI Equalized - Red Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.subplot(4,2,6)
plt.hist(rgb_hsi_eq[:,:,1].ravel(), bins=256, color='green')
plt.title("HSI Equalized - Green Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.subplot(4,2,8)
plt.hist(rgb_hsi_eq[:,:,2].ravel(), bins=256, color='blue')
plt.title("HSI Equalized - Blue Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
